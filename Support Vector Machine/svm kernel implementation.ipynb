## SVM Kernels indepth Intuition & Practical Implementation

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.linspace(-5, 5, 100)
y = np.sqrt(10**2 - x**2)  # equation of a circle with radius 10
y = np.hstack((y, -y))  # include both the upper and lower parts of the circle
x = np.hstack((x, -x))  # duplicate x values for both parts of the circle
```

```python
x1 = np.linspace(-5, 5, 100)
y1 = np.sqrt(5**2 - x1**2)  # equation of a circle with radius 5
y1 = np.hstack((y1, -y1))  # include both the upper and lower parts of the circle
x1 = np.hstack((x1, -x1))  # duplicate x values for both parts of the circle
```

```python
plt.scatter(y, x)
plt.scatter(y1, x1)
plt.xlabel('y')
plt.ylabel('x')
plt.title('Data points in 2D space')
plt.grid()
plt.show()
```
